```python
from pyspark.sql import SparkSession
from pyspark.ml.classification import LogisticRegression
from pyspark.ml.feature import VectorAssembler
from pyspark.ml.linalg import Vectors
from pyspark.sql.functions import rand

spark = SparkSession.builder.appName("Synthetic Data Classification").getOrCreate()

# Generate synthetic data
data = spark.range(1000).withColumn("feature1", rand()).withColumn("feature2", rand()).withColumn("feature3", rand()).withColumn("feature4", rand()).withColumn("feature5", rand()).withColumn("label", (rand() > 0.5).cast("int"))

# Assemble features
assembler = VectorAssembler(inputCols=["feature1", "feature2", "feature3", "feature4", "feature5"], outputCol="features")
data = assembler.transform(data)

# Logistic Regression model
lr = LogisticRegression(featuresCol="features", labelCol="label")
```

```python
model = lr.fit(data)

# Predictions
predictions = model.transform(data)
predictions.show()
```
